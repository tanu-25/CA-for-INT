```python
y=x
from sklearn.model_selection import train_test_split  
train_x,test_x,train_y,test_y=train_test_split(x,y,test_size=0.3)
```

```python
train_x.shape
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(train_x,train_y)
train_pred_lr=lr.predict(train_x)
test_pred_lr=lr.predict(test_x)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
print("Linear accuracy:",r2_score(train_pred_lr,train_y))
print("Linear accuracy:",r2_score(test_pred_lr,test_y))
```

```python
from sklearn.tree import DecisionTreeRegressor
dec=DecisionTreeRegressor()
dec.fit(train_x,train_y)
train_pred_dec=dec.predict(train_x)
test_pred_dec=dec.predict(test_x)
```

```python
print("Decision accuracy:",r2_score(train_pred_dec,train_y))
print("Decision accuracy:",r2_score(test_pred_dec,test_y))
```

```python
from sklearn.svm import SVR
svr=SVR()
svr.fit(train_x,train_y)
train_pred_svr= svr.predict(train_x)
test_pred__svr=svr.predict(test_x)
```
